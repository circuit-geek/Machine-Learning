```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
data = pd.read_csv('data.csv')
X = data.iloc[:, 0]
y = data.iloc[:, 1]
plt.scatter(X,y)
```

```python
m = 0
c = 0

L = 0.0001
epochs = 2000

n = float(len(X))
for i in range(epochs):
    y_pred = m*X + c
    D_m = (-2/n) * sum(X*(y - y_pred))
    D_c = (-2/n) * sum(y - y_pred)
    m = m - L * D_m
    c = c - L * D_c
    print(m, c, D_m)
```

```python
y_pred = m*X + c

plt.scatter(X, y)
plt.plot([min(X), max(X)], [min(y_pred), max(y_pred)], color = 'red')
```
